```python
import pandas as pd
df = pd.read_csv("5year.csv")
df.head(5)
```

```python
df[0:]
```

```python
df["score1"]=df["比分"].str[0:1]
df["score2"]=df["比分"].str[2:3]
df["judge"]=df["score1"]
```

```python
index=[]
for i in range(1930):
    if(df['比分'][i]=='	   		延期' or df['主队指数'][i]=='-'):
        index.append(i)
        print(i)
df=df.drop(index)
```

```python
df
```

```python
def change(x):
    if(x=='-'):
        return None
    else:
        return float(x)
```

```python
df['主队指数']=df['主队指数'].apply(lambda x:change(x))
df['客队指数']=df['客队指数'].apply(lambda x:change(x))
df['平指数']=df['平指数'].apply(lambda x:change(x))
```

```python
df.to_csv("5year2.csv",index=False)
```
